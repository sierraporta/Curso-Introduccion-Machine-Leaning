# Aprendizaje no supervisado

El aprendizaje no supervisado es una rama del aprendizaje automático (machine learning) en la que un algoritmo se entrena para encontrar patrones o estructuras ocultas en un conjunto de datos sin la ayuda de etiquetas o respuestas previas. A diferencia del aprendizaje supervisado, donde los datos de entrenamiento están etiquetados y se proporcionan respuestas conocidas, en el aprendizaje no supervisado no se brinda ninguna guía explícita sobre las salidas esperadas.

En el aprendizaje no supervisado, el algoritmo busca descubrir patrones o agrupaciones inherentes en los datos sin conocer de antemano cuáles son. Los métodos de aprendizaje no supervisado se utilizan a menudo cuando se dispone de grandes cantidades de datos no etiquetados y se busca obtener una comprensión más profunda de los mismos.

Los algoritmos de aprendizaje no supervisado se pueden utilizar para realizar tareas como clustering (agrupación de datos similares), detección de anomalías (identificación de patrones inusuales), reducción de la dimensionalidad (representación más compacta de los datos) y generación de datos (crear nuevas muestras similares a las existentes).

En resumen, el aprendizaje no supervisado permite descubrir patrones y estructuras ocultas en los datos sin necesidad de etiquetas previas, lo que puede conducir a un mejor entendimiento de los datos y la generación de información útil para futuros análisis o aplicaciones.

## El aprendizaje no supervisado:
se refiere a un enfoque de aprendizaje automático donde no se utilizan datos "etiquetados" para el entrenamiento. En lugar de ello, solo se tienen datos de entrada sin datos de salida correspondientes. El objetivo principal es describir la estructura de los datos y encontrar posibles patrones u organizaciones que simplifiquen el análisis. Por lo tanto, este tipo de aprendizaje tiende a ser exploratorio.

Por ejemplo, en tareas de clustering, se busca agrupar los datos en base a sus similitudes, aunque no se garantiza que estos grupos tengan un significado o utilidad específica. En ocasiones, al explorar los datos sin un objetivo definido, es posible encontrar correlaciones curiosas pero poco prácticas.

El aprendizaje no supervisado se utiliza comúnmente en:

- Problemas de clustering
- Agrupamientos de co-ocurrencias
- Perfilado o profiling.

Sin embargo, es importante mencionar que problemas que involucran tareas como encontrar similitud, predicción de enlaces o reducción de datos pueden ser abordados tanto con enfoques supervisados como no supervisados.

Algunos de los algoritmos más utilizados en el aprendizaje no supervisado son:

1. Algoritmos de clustering
2. Análisis de componentes principales
3. Descomposición en valores singulares (Singular Value Decomposition)
4. Análisis de componentes independientes (Independent Component Analysis)

## Diferentes tipos de preprocesamiento, Aplicación de transformaciones de datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
df["diagnostic"]=cancer.target
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
# transform data
X_train_scaled = scaler.transform(X_train)
# print dataset properties before and after scaling
print("transformed shape: {}".format(X_train_scaled.shape))
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(
    X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
    X_train_scaled.max(axis=0)))
```

```python
# transform test data
X_test_scaled = scaler.transform(X_test)
# print test data properties after scaling
print("per-feature minimum after scaling:\n{}".format(X_test_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n{}".format(X_test_scaled.max(axis=0)))
```

# Using UMAP for Clustering
UMAP can be used as an effective preprocessing step to boost the performance of density based clustering. This is somewhat controversial, and should be attempted with care. For a good discussion of some of the issues involved in this, please see the various answers in this stackoverflow thread on clustering the results of t-SNE. Many of the points of concern raised there are salient for clustering the results of UMAP. The most notable is that UMAP, like t-SNE, does not completely preserve density. UMAP, like t-SNE, can also create false tears in clusters, resulting in a finer clustering than is necessarily present in the data. Despite these concerns there are still valid reasons to use UMAP as a preprocessing step for clustering. As with any clustering approach one will want to do some exploration and evaluation of the clusters that come out to try to validate them if possible.

```python
from sklearn.decomposition import PCA

# Dimension reduction and clustering libraries
import umap
#import hdbscan
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score
```

```python
standard_embedding = umap.UMAP(random_state=42).fit_transform(cancer.data)
plt.scatter(standard_embedding[:, 0], standard_embedding[:, 1], c=cancer.target, s=20, cmap='Spectral');
```

# Traditional clustering
Now we would like to cluster the data. As a first attempt let’s try the traditional approach: K-Means. In this case we can solve one of the hard problems for K-Means clustering – choosing the right k value, giving the number of clusters we are looking for. In this case we know the answer is exactly 10. We will use sklearns K-Means implementation looking for 10 clusters in the original 784 dimensional data.

```python
import sklearn.cluster as cluster

kmeans_labels = cluster.KMeans(n_clusters=2).fit_predict(cancer.data)
```

```python
plt.scatter(standard_embedding[:, 0], standard_embedding[:, 1], c=kmeans_labels, s=20, cmap='Spectral');
```

# Using UMAP for Clustering, More...........

```python
from sklearn.preprocessing import QuantileTransformer, FunctionTransformer, PowerTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

# Preprocess again
pipe = make_pipeline(SimpleImputer(strategy="mean"), QuantileTransformer())
X = pipe.fit_transform(cancer.data.copy())

# Fit UMAP to processed data
manifold = umap.UMAP(random_state=42).fit(X)
X_reduced_2 = manifold.transform(X)
```

```python
plt.scatter(X_reduced_2[:, 0], X_reduced_2[:, 1], c=cancer.target, s=20, cmap='Spectral');
```
